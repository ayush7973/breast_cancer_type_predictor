```
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```
cancer=pd.read_csv("/content/breast_cancer_data.csv")
```

```
cancer.head()
```

```
df=pd.DataFrame(cancer)
check= df.isnull().values.any()
```

```
print(check)
```

```
column_name=df.columns[df.isnull().any()].tolist()
```

```
print(column_name
      )
```

```
cancer=df.drop(columns=['Unnamed: 32'])
```

```
cancer.head()
```

```
diagnosis=cancer['diagnosis']
```

```
others=cancer.drop(columns=['diagnosis'])
```

```
diagnosis.head()
```

```
others.head()
```

```
scaler=StandardScaler()
```

```
scaler.fit(others)
```

```
sddata=scaler.transform(others)
```

```
print(sddata)
```

```
df=pd.DataFrame(sddata)
check= df.isnull().values.any()
print(check)
```

```
x=sddata
y=diagnosis
```

```
x_train, x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```
print(x.shape, x_train.shape,x_test.shape)
```

```
classifier=svm.SVC(kernel='linear')
```

```
classifier.fit(x_train,y_train)
```

```
x_train_prediction=classifier.predict(x_train)
trainingaccuracy=accuracy_score(x_train_prediction,y_train)
```

```
print(trainingaccuracy)
```

```
x_test_prediction=classifier.predict(x_test)
testaccuracy=accuracy_score(x_test_prediction,y_test)
```

```
print(testaccuracy)
```
